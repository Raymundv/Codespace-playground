Data analysis for Diamonds data set

```python
#importing necessary libraries
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
diamonds_data = sns.load_dataset('diamonds')
```

```python
diamonds_data.head()
```

```python
diamonds_data.describe()
```

Creating visualizations

Investigating correlation of variables (Visually)

```python
#setting the style of the plots to be used
sns.set_style('whitegrid')

#creating the relational plot using the seaborn library, (rel plot). 

g = sns.relplot(x='carat', y='price', data=diamonds_data, kind='scatter', height=5, aspect=2)
g.figure.suptitle('Price vs carat')
g.set_axis_labels('Carat', 'Price')
```

Distributions for numerical variables

```python
g = sns.displot(data=diamonds_data, x='price', kind='hist',kde=True, height=5, aspect=2)
g.fig.suptitle("Price distribution")
g.set_axis_labels('Price', 'Frequency') 
```

```python
g = sns.histplot(data=diamonds_data, x='carat', kde=True)
g.set_title('Carat distribution')
g.set_xlabel('Carat')
g.set_ylabel('Frequency')
```

Numerical variables grouped by category

```python
g = sns.catplot(data=diamonds_data, x='cut',y='price', kind='box', height=5, aspect=2, order=['Fair', 'Good', 'Very Good', 'Premium', 'Ideal'])
g.fig.suptitle('Price vs Cut', va ='baseline', ha = 'center')
g.set_axis_labels('Cut', 'Price')
```

Creating a jointplot

```python
g = sns.jointplot(data=diamonds_data, x = 'depth', y = 'price', kind = 'kde', height = 5, ratio = 5,fill=True,space = 0)
```

```python
#Creating a pairplot to show the relationship between the numerical columns in the dataset
g = sns.pairplot(data=diamonds_data, height=5, aspect=2, corner=True)
g.figure.suptitle('Pairplot of the diamonds dataset')
```
